```python
import pandas as pd

# Load dataset
df = pd.read_csv("data_cleaning_sample.csv")

# 1. Check missing values
df.isnull().sum()

# 2. Convert Join Date to datetime
df["Join Date"] = pd.to_datetime(df["Join Date"], errors="coerce")

# 3. Fill missing Age with mean
df["Age"] = df["Age"].fillna(df["Age"].mean())

# 4. Fill missing City with mode
df["City"] = df["City"].fillna(df["City"].mode()[0])

# 5. Fill missing Gender with mode
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])

# 6. Fill missing Email with placeholder
df["Email"] = df["Email"].fillna("not_provided")

# 7. Fill missing Join Date with median date
df["Join Date"] = df["Join Date"].fillna(df["Join Date"].median())

# 8. Remove duplicate rows (if any)
df = df.drop_duplicates()

# 9. Reset index
df.reset_index(drop=True, inplace=True)

# 10. Final check
df.info()
```
