```python
import pandas as pd
```

* Series = 1D array with labels

```python
s1 = pd.Series([71, 84, 56, 23, 56, 98, 57])
print(s1)
```

```python
type(s1)
```

```python
s2 = pd.Series([91, 84, 56, 23, 56, 98, 57], ["Aditya", "Shubh", "Rohan", "Aakash", "Kirti", "John", "Rehan"])
print(s2)
```

```python
print(s2["Aditya"])
```

* DataFrame = 2D table with rows + columns

```python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["Delhi", "Mumbai", "Bangalore"]
}
df = pd.DataFrame(data)
print(df)
```
