```python
import pandas as pd
```

                                 DataFrame using List

```python
data = [['Aditya', 94], ['harry', 84], ['Jill', 74], ['Jacob', 64], ['jack', 54], ['Jerry', 44]]
print(data)
```

```python
df = pd.DataFrame(data, columns=["Name", "Marks"])
print(df)
```

                                 DataFrame using dict

```python
data = {
    "a": [4, 6, 3],
    "b": [56, 34, 78]
}
print(data)
```

```python
df1 = pd.DataFrame(data)
print(df1)
```

                                  DataFrame using Numpy

```python
import numpy as np
```

```python
arr = np.array([[1, 2], [3, 4]])
ndf = pd.DataFrame(arr, columns=["A", "B"])
print(ndf)
```
