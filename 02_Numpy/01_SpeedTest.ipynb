```python
# using python lists

import numpy as np
import time

size = 1_000_000
l1 = list(range(size))
l2 = list(range(size))

start = time.time()   # note start time
add = [x+y for x,y in zip(l1, l2)]
end = time.time()   # note end time

print(add[0:10])
print(end - start)
```

```python
# using numpy array

import numpy as np
import time

size = 1_000_000
l1 = np.array(range(size))
l2 = np.array(range(size))

start = time.time()   # note start time
add = l1 + l2

end = time.time()   # note end time

print(add[0:10])
print(end - start)
```

```python
import sys
 
list_data = list(range(1000))
numpy_data = np.array(list_data)
 
print("Python list size:", sys.getsizeof(list_data) * len(list_data), "bytes")
print("NumPy array size:", numpy_data.nbytes, "bytes")
```

```python
# Square of elements using lists
size = 1_000_000
li = list(range(size))

start = time.time()
list_square = [x**2 for x in li]
end = time.time()

print(end - start)
```

```python
# Square of elements using numpy
size = 1_000_000
arr1 = np.array(range(size))

start = time.time()
numpy_square = arr1 ** 2
end = time.time()

print(end - start)
```
